```python
import os
import pandas as pd
import json
import random
import shutil
```

```python
#splitting images on apples category into training and val set
```

```python
apples_images = "../Fruits/dataset/apples/images (copy)"
train = "../Fruits/dataset/apples/train"
val = "../Fruits/dataset/apples/val"
```

```python
images = os.listdir(apples_images)
```

```python
t = random.sample(images,int(len(images)/6))
```

```python
for image in t:
    source = os.path.join(apples_images,image)
    destination = os.path.join(val,image)
    shutil.move(source,destination)
```

```python
#splitting csv annotations into train and val
```

```python
annotations = "../Fruits/dataset/apples/annotations_csv (copy)"
val_csv = "../Fruits/dataset/apples/val_csv"
train_csv = "../Fruits/dataset/apples/train_csv"
```

```python
#image name will determine the name of the csv file
image_names = os.listdir(val)

for item in image_names:
    csv_name = item.replace(".png",".csv")
    source = os.path.join(annotations,csv_name)
    destination = os.path.join(val_csv,csv_name)
    shutil.move(source,destination)
```

```python
#remove extra viles on the train csv
image_names = os.listdir(train)

for item in image_names:
    csv_name = item.replace(".png",".csv")
    file = os.path.join(annotations,csv_name)
    source = os.path.join(annotations,csv_name)
    destination = os.path.join(train_csv,csv_name)
    shutil.move(source,destination)
```
