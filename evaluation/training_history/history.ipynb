```python
import seaborn as sns
import json
import os
```

```python
history_files = [i for i in os.listdir("history") if i.endswith(".json")]
```

```python
exp = json.load(open(os.path.join("history", history_files[0])))
```

```python
exp = [i for i in exp if i["tag"]=="val_loss"]
```

```python
steps = [i["epoch"] for i in exp]
```

```python
values = [i["step_value"] for i in exp]
```

```python
sns.lineplot(x=steps, y=values)
```
